# Testing Stationarity of Stocks 

##  Checking constant mean and variance

```python
pd.DataFrame()
```

```python
def get_mean_var(series, no_of_samples = 10):
    split_size = int(len(series) / no_of_samples)
    start = 0
    mean_var_list = []
    for i in range(no_of_samples):
        sample_series = series[i*split_size:(i+1)*split_size]
        #print(sample_series,"\n")
        #print('Mean= %.2f, Variance= %.2f' % (sample_series.mean(), sample_series.var()))
        mean_var_list.append({'mean':round(sample_series.mean(),2), 'variance':round(sample_series.var(),2)})
    return pd.DataFrame(mean_var_list)
```

```python
mean_var_df = pd.concat([get_mean_var(df[stock]) for stock in stocks],axis = 1,keys = stocks)
mean_var_df
for i,stock in enumerate(mean_var_df):
    if i%2 == 0:
        fig = plt.figure(figsize = (30,10))
        ax1 = fig.add_subplot(121)
        ax1.plot(mean_var_df[stock])
        ax1.set_title(stock,fontsize = 30)
        for tick in ax1.xaxis.get_major_ticks():
                tick.label.set_fontsize(25) 
        for tick in ax1.yaxis.get_major_ticks():
                tick.label.set_fontsize(25) 
        
    else:
        ax2 = fig.add_subplot(122)
        ax2.plot(mean_var_df[stock])
        ax2.set_title(stock,fontsize = 30)
        for tick in ax2.xaxis.get_major_ticks():
                tick.label.set_fontsize(25) 
        for tick in ax2.yaxis.get_major_ticks():
                tick.label.set_fontsize(25)
        plt.show()
```

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics   (daily 1.30pm to 8pm) Use Daily because some weeks are 4 days some 5 days, irregular
    rolmean = timeseries.rolling(window = 390).mean()  
    rolstd = timeseries.rolling(window = 390).std()
    
    #Plot rolling statistics:
    plt.figure(figsize=(20,10))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title(f'Rolling Mean & Standard Deviation of {timeseries.name}')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
    return dfoutput
```

```python
df.apply(test_stationarity)
```

None of the Stocks are stationary currently,all their adfuller test statistic (null hypothesis is non-stationary) are more than even the Critical Value at 10%

# Estimating & Eliminating Trend 

## Generating Moving Averages 

```python
df_movingave = pd.DataFrame()
for col in df.columns:
    df_movingave[col] = df[col].rolling(window = 390).mean()
```

```python
df_movingave_diff = df - df_movingave
df_movingave_diff.dropna(inplace=True)
```

# Testing Stationarity after trend removal by Moving Average 

```python
df_movingave_diff.apply(test_stationarity)
```

```python
for stock in stocks:
    durbin_watson(df_movingave_diff[stock])
```

Judging by their test statistic being less than the 1% critical value, there is an improvement from the more than 10% critical value previously before removing trend by moving average. However the DW statistic is still not close to 2.

Further experiments can be done with 
-  differencing https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.shift.html 
-  decomposition https://www.statsmodels.org/dev/generated/statsmodels.tsa.seasonal.seasonal_decompose.html

# Eliminating Trend and Seasonality with Differencing

```python
df_shifted = pd.DataFrame()
for col in df.columns:
    df_shifted[col] = df[col].shift()
```

```python
df_diff = df-df_shifted
df_diff.dropna(inplace=True)
```

# Testing Stationarity after trend and seasonality removal by differencing 

```python
df_diff.apply(test_stationarity)
```

```python
for stock in stocks:
    durbin_watson(df_diff[stock])
```

The first differencing test statistic is significantly better than trend removal by Moving average, we will use this series going forward. Its also easier to add noise and seasonality back into predicted residuals with differencing method. The DW test also shows values close to 2

## Rules of thumb 

ARIMA(p,d,q)

* **Number of AR (Auto-Regressive) terms (p)**: AR terms are just lags of dependent variable. For instance if p is 5, the predictors for x(t) will be x(t-1)….x(t-5).
* **Number of MA (Moving Average) terms (q)**: MA terms are lagged forecast errors in prediction equation. For instance if q is 5, the predictors for x(t) will be e(t-1)….e(t-5) where e(i) is the difference between the moving average at ith instant and actual value.
* **Number of Differences (d)**: These are the number of nonseasonal differences, i.e. in this case we took the first order difference. So either we can pass that variable and put d=0 or pass the original variable and put d=1. Both will generate same results.

* p – The lag value where the PACF chart crosses the upper confidence interval for the first time. 
* q – The lag value where the ACF chart crosses the upper confidence interval for the first time.

```python
durbin_watson(results_ARIMA.resid)
```

```python
from statsmodels.graphics.api import qqplot

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
fig = qqplot(results_ARIMA.resid, line='q', ax=ax, fit=True)
```

### Adding it back to the original scale

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()

predictions_ARIMA = pd.Series(df['NASDAQ.AAPL'].iloc[0], index=df['NASDAQ.AAPL'].index)
predictions_ARIMA = predictions_ARIMA.add(predictions_ARIMA_diff_cumsum,fill_value=0)

plt.plot(df['NASDAQ.AAPL'])
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-df['NASDAQ.AAPL'])**2)/len(df['NASDAQ.AAPL'])))
```
